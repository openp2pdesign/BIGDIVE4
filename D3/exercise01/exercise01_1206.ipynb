```python
# Add NaN columns for individual
data["individual_gold"] = np.nan
data["individual_silver"] = np.nan
data["individual_bronze"] = np.nan
```

```python
def isint(x):
    try:
        a = float(x)
        b = int(a)
    except ValueError:
        return False
    else:
        return a == b
```

```python
for i in data.index:
    # Check only who won a medal
    print ""
    print i
    if data.ix[i]["Medals"] == "YES":
        # Gold
        # Check for value: NaN, int (individual), float (team)
        if str(data.ix[i]["Gold"]) != "nan":
            value1 = data.ix[i]["Gold"]
            print "_GOLD",value1,type(value1)

            # Check if int (individual) or float (team) 
            if isint(value1):
                print data.ix[i]["Name"],"__individual__GOLD", data.ix[i]["Event"],"__",data.ix[i]["Sport"]
                data.ix[i,"individual_gold"] = True
            else:
                print data.ix[i]["Name"],"__team__GOLD", data.ix[i]["Event"],"__",data.ix[i]["Sport"]
                data.ix[i,"individual_gold"] = False
        
        # Silver
        # Check for value: NaN, int (individual), float (team)
        if str(data.ix[i]["Silver"]) != "nan":
            
            value2 = data.ix[i]["Silver"]
            print "_Silver",value2,type(value2)

            # Check if int (individual) or float (team) 
            if isint(value2):
                print data.ix[i]["Name"],"__individual__SILVER", data.ix[i]["Event"],"__",data.ix[i]["Sport"]
                data.ix[i,"individual_silver"] = True
            else:
                print data.ix[i]["Name"],"__team__SILVEr", data.ix[i]["Event"],"__",data.ix[i]["Sport"]
                data.ix[i,"individual_silver"] = False
                    
        # Bronze
        # Check for value: NaN, int (individual), float (team)
        if str(data.ix[i]["Bronze"]) != "nan":
            
            value3 = data.ix[i]["Bronze"]
            print "_BRONZE",value3,type(value3)

            # Check if int (individual) or float (team) 
            if isint(value3):
                print data.ix[i]["Name"],"__individual__BRONZE", data.ix[i]["Event"],"__",data.ix[i]["Sport"]
                data.ix[i,"individual_bronze"] = True
            else:
                print data.ix[i]["Name"],"__team__BRONZE", data.ix[i]["Event"],"__",data.ix[i]["Sport"]
                data.ix[i,"individual_bronze"] = False
```
